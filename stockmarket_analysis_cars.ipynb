## Read data from multiple car companies 

```python
tesla = web.DataReader(name="TSLA",data_source="yahoo",start=start,end=end)
nissan = web.DataReader(name="NSANY",data_source="yahoo",start=start,end=end)
bmw = web.DataReader(name="BMW.DE",data_source="yahoo",start=start,end=end)
mercedez = web.DataReader(name="DAI.DE",data_source="yahoo",start=start,end=end)
ford = web.DataReader(name="F",data_source="yahoo",start=start,end=end)
toyota = web.DataReader(name="TM",data_source="yahoo",start=start,end=end)
vw = web.DataReader(name="VWAGY",data_source="yahoo",start=start,end=end)
```

```python
plt.figure(figsize=(16,5))
#tesla["Close"].plot(label="Tesla",figsize=(16,5))
nissan["Close"].plot(label="Nissan")
bmw["Close"].plot(label="BMW")
mercedez["Close"].plot(label="Mercedez")
ford["Close"].plot(label="Ford")
toyota["Close"].plot(label="Toyota")
vw["Close"].plot(label="Volkswagon")
plt.legend()
plt.title("Close prices for all cars")
plt.show()
```

```python
plt.figure(figsize=(16,5))
#tesla["Close"].plot(label="Tesla",figsize=(16,5))
nissan["Open"].plot(label="Nissan")
bmw["Open"].plot(label="BMW")
mercedez["Open"].plot(label="Mercedez")
ford["Open"].plot(label="Ford")
toyota["Open"].plot(label="Toyota")
vw["Open"].plot(label="Volkswagon")
plt.legend()
plt.title("Open prices for all cars")
plt.show()
```

```python
plt.figure(figsize=(16,5))
#tesla["Close"].plot(label="Tesla")
nissan["Volume"].plot(label="Nissan")
bmw["Volume"].plot(label="BMW")
mercedez["Volume"].plot(label="Mercedez")
ford["Volume"].plot(label="Ford")
toyota["Volume"].plot(label="Toyota")
vw["Volume"].plot(label="Volkswagon")
plt.legend()
plt.title("Volume for all cars")
plt.show()
```

```python
tesla["total_traded"] = tesla["Open"]*tesla["Volume"]
nissan["total_traded"] = nissan["Open"]*nissan["Volume"]
bmw["total_traded"] = bmw["Open"]*bmw["Volume"]
mercedez["total_traded"] = mercedez["Open"]*mercedez["Volume"]
ford["total_traded"] = ford["Open"]*ford["Volume"]
toyota["total_traded"] = toyota["Open"]*toyota["Volume"]
vw["total_traded"] = vw["Open"]*vw["Volume"]
```

```python
plt.figure(figsize=(16,5))
#tesla["Close"].plot(label="Tesla")
nissan["total_traded"].plot(label="Nissan")
bmw["total_traded"].plot(label="BMW")
mercedez["total_traded"].plot(label="Mercedez")
ford["total_traded"].plot(label="Ford")
toyota["total_traded"].plot(label="Toyota")
vw["total_traded"].plot(label="Volkswagon")
plt.legend()
plt.title("Total traded for all cars")
plt.show()
```

## Moving Average

```python
ford["Close"].plot(label="Close",figsize=(16,5))
ford['ma60'] = ford["Close"].rolling(60).mean() #rolling window calculations
ford['ma60'].plot(label="MA60")

ford['ma300'] = ford["Close"].rolling(300).mean() #rolling window calculations
ford['ma300'].plot(label="MA300")

plt.legend()
plt.title("Ford Close moving avg")
plt.show()
```

## Analyze Returns

```python
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) - 1
nissan['returns'] = (nissan['Close']/nissan['Close'].shift(1)) - 1
bmw['returns'] = (bmw['Close']/bmw['Close'].shift(1)) - 1
mercedez['returns'] = (mercedez['Close']/mercedez['Close'].shift(1)) - 1
ford['returns'] = (ford['Close']/ford['Close'].shift(1)) - 1
toyota['returns'] = (toyota['Close']/toyota['Close'].shift(1)) - 1
vw['returns'] = (vw['Close']/vw['Close'].shift(1)) - 1
```

```python
tesla['returns'].hist(bins=50)
```

```python
nissan['returns'].hist(bins=50)
```

```python
toyota['returns'].hist(bins=50)
```

## Analyze cumulative returns 

```python
tesla['Cumulative Return'] = (1+tesla['returns']).cumprod()
nissan['Cumulative Return'] = (1+nissan['returns']).cumprod()
bmw['Cumulative Return'] = (1+bmw['returns']).cumprod()
mercedez['Cumulative Return'] = (1+mercedez['returns']).cumprod()
ford['Cumulative Return'] = (1+ford['returns']).cumprod()
toyota['Cumulative Return'] = (1+toyota['returns']).cumprod()
vw['Cumulative Return'] = (1+vw['returns']).cumprod()
```

```python
bmw['Cumulative Return'].plot(label='BMW', figsize=(16,5))
vw['Cumulative Return'].plot(label='VW')
mercedez['Cumulative Return'].plot(label='Mercedez')
toyota['Cumulative Return'].plot(label='Toyota')
nissan['Cumulative Return'].plot(label='Nissan')
#tesla['Cumulative Return'].plot(label='Tesla')
ford['Cumulative Return'].plot(label='Ford')
plt.title('Cumulative Return Vs Time')
plt.legend()
```
